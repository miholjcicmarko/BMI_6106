```r
meanconfint = function (x, sd, level) {
  len = as.numeric(length(x))
  se = sd / sqrt(len)
  mu = mean(x)
  z = qnorm(1 - (1 - level)/2)
  out = c(mu - z * se, " to ", mu + z * se)
  names(out) = c("lower", "upper")
  return(out)
}
```

### B. 

A study looked at various cardiovascular risk factors in children, as measured at birth and during their first five years of life. One of the results for newborns were to study differences in bits per minute from two different races.
Based on the data, Is there evidence that the number of heart beats/min of newborn white children is fewer than that of newborn black children? 

The data is located in this working directory with the name Cardiovascular_children.csv

The null hypothesis is that the number of heart beats/min of newborn white children is the same as that of newborn black children.
The alternative hypothesis is that the number of heart beats/min of newborn white children is fewer than that of newborn black children.
The test is a one-tailed test. 

```r
df = read.csv("Cardiovascular_children.csv", header = T)
df
```

```r
tab = table(df$race)
tab
```

```r
cohen.ES(test = "t", size = "small")
pwr.t2n.test(n1 = 25, n2 = 121, d = 0.2)
```

This test has very small power at 14.765%. This means that there is a 14.675% chance of correctly rejecting the null hypothesis at the 0.05 significance level using this test. 

```r
pwr.t.test(d = 0.2, power = 0.80, sig.level = 0.05)
```

This test tells us that 394 samples in each group are needed in order to have a test with a power of 80%. This means that there would be a 80% chance of correctly rejecting the null hypothesis at the 0.05 significance level using this test with 394 samples each.

```r
white = df[which(df$race == "white"), ]
black = df[which(df$race == "black"), ]

p = replicate(10000, t.test(white$bpm, black$bpm)$p.value)
p_val = mean(p)
p_val
```

Mean value of p value is 1.17750794330602e-06. 

```r
ttest = t.test(white$bpm, black$bpm, alternative = "less", paired = FALSE)
ttest
```

This t-test determined that there is a difference between the two groups. The p-value is well below 0.05 and the confidence interval consists of negative numbers. Therefore, the test states that the white sample has a smaller BPM mean than the black sample. We can accept our alternative hypothesis. The number of heart beats/min of newborn white children is fewer than that of newborn black children. There are some reservations to this conclusion because the power of this analysis was not very strong.

```r
meanconfintrep = function (x, sd, level) {
  len = as.numeric(length(x))
  se = sd / sqrt(len)
  mu = mean(x)
  z = qnorm(1 - (1 - level)/2)
  out = c((mu + z * se), " to ", (mu - z * se))
  names(out) = c("lower", "upper")
  return(out)
}
```

```r
confintO <- meanconfint((white$bpm), sd((white$bpm)), 0.95)
confintOW <- meanconfint((white2$bpm), sd((white2$bpm)), 0.95)
cat("Mean 95% Confidence Interval for White Sample with outliers: ", (confintO), " \n")
cat("Mean 95% Confidence Interval for White Sample without outliers: ", (confintOW))
```

```r
confintOWT <- meanconfintrep(1/(white2$bpm), sd(1/(white2$bpm)), 0.95)
cat("Mean 95% Confidence Interval for Reciprocal Transform of White Sample without outliers: ", (confintOWT))
```

```r
confintO <- meanconfint(black$bpm, sd(black$bpm), 0.95)
confintOW <- meanconfint(black2$bpm, sd(black2$bpm) , 0.95)
cat("Mean 95% Confidence Interval for Black Sample with outlier: ", confintO, " \n")
cat("Mean 95% Confidence Interval for Black Sample without outlier: ", confintOW)
```

The confidence intervals provide evidence that the t-test conclusion was correct. This is because the confidence intervals of the two groups do not overlap each other.

Among females in the US between the ages of 18 and 74, diastolic blood pressure is normally distributed with mean 77 mm Hg and standard deviation of 11.6 mm Hg.  What is the probability that a randomly selected woman has a diastolic blood pressure less than 60 mm Hg?  greater than 90 mm Hg? between 60 and 90 mm Hg?

	X ~ N(77, 134.56)

```r
prob = pnorm(60,mean = 77,sd = sqrt(134.56))
cat("probability that a randomly selected woman has a diastolic blood pressure less than 60 mm Hg: ", prob, "\n")
```

```r
mean <- 77
std <- sqrt(134.56)
females_in_US <- 166745678
error <- qnorm(0.975)*std/sqrt(females_in_US)
left <- mean-error
right <- mean+error
cat("Mean 95% Confidence Interval for Mean Diastolic Blood Pressure: ", left, " ", right, " \n")
```

```r
prob = 1 - pnorm(90,mean = 77,sd = sqrt(134.56))
cat("probability that a randomly selected woman has a diastolic blood pressure greater than 90 mm Hg: ", prob, "\n")
```

```r
prob = (pnorm(90,mean = 77,sd = sqrt(134.56))) - (pnorm(60,mean = 77,sd = sqrt(134.56)))
cat("probability that a randomly selected woman has a diastolic blood pressure greater than 60 mm Hg and lower than 90 mm Hg: ", prob, "\n")
```
