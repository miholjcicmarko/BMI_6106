```r
#PCA
library(tidyverse)
scaled_df <- apply(calls, 2, scale)

call.cov <- cov(scaled_df)
call.eigen <- eigen(call.cov)

phi <- call.eigen$vectors[,1:2]
row.names(phi) <- c("rate_high_edu", "rate_call_911", "rate_emp", "rate_Race_w", "income")
colnames(phi) <- c("PC1", "PC2")
phi
```

```r
PC1 <- as.matrix(scaled_df) %*% phi[,1]
PC2 <- as.matrix(scaled_df) %*% phi[,2]

# Create data frame with Principal Components scores
PC <- data.frame(Town = row.names(calls), PC1, PC2)
head(PC)
```

```r
##Create biplot
library(ggplot2)
# Plot Principal Components for each State
ggplot(PC, aes(PC1, PC2)) + 
  modelr::geom_ref_line(h = 0) +
  modelr::geom_ref_line(v = 0) +
  geom_text(aes(label = Town), size = 2) +
  xlab("First Principal Component") + 
  ylab("Second Principal Component") + 
  ggtitle("First Two Principal Components of Calls Data")
```

```r
library(gridExtra)
```

```r
##Scree plot
PVE <- call.eigen$values / sum(call.eigen$values)

# PVE (aka scree) plot
PVEplot <- qplot(c(1:5), PVE) + 
  geom_line() + 
  xlab("Principal Component") + 
  ylab("PVE") +
  ggtitle("Scree Plot") +
  ylim(0, 1)

# Cumulative PVE plot
cumPVE <- qplot(c(1:5), cumsum(PVE)) + 
  geom_line() + 
  xlab("Principal Component") + 
  ylab(NULL) + 
  ggtitle("Cumulative Scree Plot") +
  ylim(0,1)

grid.arrange(PVEplot, cumPVE, ncol = 2)
```

```r
##Loadings table
loadings <- call.eigen$vectors[,1:2] * sqrt(call.eigen$values)

phi <- loadings
row.names(phi) <- c("rate_high_edu", "rate_call_911", "rate_emp", "rate_Race_w", "income")
colnames(phi) <- c("PC1", "PC2")
phi
```

Close to 75% of the variance is explained by the first two principal components. The third principal component was the final component that made a large impact on the percentage variance explained. The fourth and fifth components made very little difference in the percent of variance explained. Rate of high education has a large effect on the first principal component. Rate of employment has a large effect on the second principal component.

```r
##K-means
##How many clusters?
library(tidyverse)  
library(cluster)    
library(factoextra) 

df <- calls
df <- na.omit(df)
df <- scale(df)
```

```r
set.seed(123)

# function to compute total within-cluster sum of square 
wss <- function(k) {
  kmeans(df, k, nstart = 10 )$tot.withinss
}

# Compute and plot wss for k = 1 to k = 15
k.values <- 1:10

# extract wss for 2-15 clusters
wss_values <- map_dbl(k.values, wss)

plot(k.values, wss_values,
       type="b", frame = FALSE, 
       xlab="Number of clusters K",
       ylab="Total within-clusters sum of squares")
```

4 clusters are needed for ideal clustering. This is because the total within-clusters sum of squares begins to gradually decrease after 4 clusters. 

```r
k4 <- kmeans(df, centers = 4, nstart = 25)
fviz_cluster(k4, data = df)
```

Ideally, there are four clusters. One of the clusters is made up of one county. This is because this county is very different from the other counties. This county does not share a large majority of characteristics found in other clusters. Therefore, it is in its own cluster. It can be considered an outlier in this dataset.
