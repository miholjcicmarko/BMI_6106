```python
import pandas as pd
import numpy as np

kidney = pd.read_csv("kidney_disease.csv")
df = pd.DataFrame(kidney)
```

```python
import math as m

def meanInserter(category, mean):
    i = 0
    
    while (i < 400):
        if (m.isnan(category[i])):
            category[i] = mean
            i += 1
        else:
            i += 1
            continue
        

    return category
```

```python
df["age"] = meanInserter(df["age"], 60)

df["bp"] = meanInserter(df["bp"], 80)

df["sg"] = meanInserter(df["sg"], 1.02)

df["al"] = meanInserter(df["al"], 0)

df["su"] = meanInserter(df["su"], 0)

df["bgr"] = meanInserter(df["bgr"], 99)

df["bu"] = meanInserter(df["bu"], 46.0)

df["sc"] = meanInserter(df["sc"], 1.2)

df["sod"] = meanInserter(df["sod"], 135)

df["pot"] = meanInserter(df["pot"], 3.5)

df["hemo"] = meanInserter(df["hemo"], 15)

for i in range(2):
    df["pcv"] = df["pcv"].replace(r'^\s*$', np.nan, regex = True)
    df["pcv"] = df["pcv"].fillna('41')
    df["pcv"] = df["pcv"].str.strip("?")
    df["pcv"] = df["pcv"].str.strip()

for i in range(2):
    df["wc"] = df["wc"].replace(r'^\s*$', np.nan, regex = True)
    df["wc"] = df["wc"].fillna('9800')
    df["wc"] = df["wc"].str.strip("?")
    df["wc"] = df["wc"].str.strip()

for i in range(2):
    df["rc"] = df["rc"].replace(r'^\s*$', np.nan, regex = True)
    df["rc"] = df["rc"].fillna('5.2')
    df["rc"] = df["rc"].str.strip("?")
    df["rc"] = df["rc"].str.strip()
```

```python
df["pcv"]= df["pcv"].astype(int)
df["wc"] = df["wc"].astype(int)
df["rc"] = df["rc"].astype(float)
df.dtypes
```

```python
filterNormal = df.loc[df["rbc"] == "normal"]
countNormal = filterNormal["rbc"].count()
print("Count of Normal for rbc is " + str(countNormal))

df["rbc"] = df["rbc"].fillna("normal")
df["rbc"] = df["rbc"].replace(r'^\s*$', np.nan, regex = True)
df["rbc"] = df["rbc"].str.strip('\t')

filterNormal_pc = df.loc[df["pc"] == "normal"]
countNormal_pc = filterNormal_pc["pc"].count()
print("Count of Normal for pc is " + str(countNormal_pc))

df["pc"] = df["pc"].fillna("normal")
df["pc"] = df["pc"].replace(r'^\s*$', np.nan, regex = True)
df["pc"] = df["pc"].str.strip('\t')

filterPresent = df.loc[df["pcc"] == "present"]
countPresent = filterPresent["pcc"].count()
print("Count of present for pcc is " + str(countPresent))

df["pcc"] = df["pcc"].fillna("notpresent")
df["pcc"] = df["pcc"].replace(r'^\s*$', np.nan, regex = True)
df["pcc"] = df["pcc"].str.strip('\t')

filterPresent_ba = df.loc[df["ba"] == "present"]
countPresent_ba = filterPresent_ba["ba"].count()
print("Count of present for ba is " + str(countPresent_ba))

df["ba"] = df["ba"].fillna("notpresent")
df["ba"] = df["ba"].replace(r'^\s*$', np.nan, regex = True)
df["ba"] = df["ba"].str.strip('\t')
```

```python
filterHtn = df.loc[df["htn"] == "yes"]
countHtnYes = filterHtn["htn"].count()
print("Count of yes for htn is " + str(countHtnYes))

#df["htn"] = df["htn"].fillna("no")
df["htn"] = df["htn"].replace(r'^\s*$', np.nan, regex = True)
df["htn"] = df["htn"].str.strip('\t')

filterDm = df.loc[df["dm"] == "yes"]
countDmYes = filterDm["dm"].count()
print("Count of yes for dm is " + str(countDmYes))

df["dm"] = df["dm"].fillna("no")
df["dm"] = df["dm"].replace(r'^\s*$', np.nan, regex = True)
df["dm"] = df["dm"].str.strip('\t')

filterCad = df.loc[df["cad"] == "yes"]
countCadYes = filterCad["cad"].count()
print("Count of yes for cad is " + str(countCadYes))

df["cad"] = df["cad"].fillna("no")
df["cad"] = df["cad"].replace(r'^\s*$', np.nan, regex = True)
df["cad"] = df["cad"].str.strip('\t')

filterAppet = df.loc[df["appet"] == "good"]
countAppet = filterAppet["appet"].count()
print("Count of good for appet is " + str(countAppet))

df["appet"] = df["appet"].fillna("good")
df["appet"] = df["appet"].replace(r'^\s*$', np.nan, regex = True)
df["appet"] = df["appet"].str.strip('\t')

filterPe = df.loc[df["pe"] == "yes"]
countPe = filterPe["pe"].count()
print("Count of yes for pe is " + str(countPe))

df["pe"] = df["pe"].fillna("no")
df["pe"] = df["pe"].replace(r'^\s*$', np.nan, regex = True)
df["pe"] = df["pe"].str.strip('\t')

filterAne = df.loc[df["ane"] == "yes"]
countAne = filterAne["ane"].count()
print("Count of yes for ane is " + str(countAne))

df["ane"] = df["ane"].fillna("no")
df["ane"] = df["ane"].replace(r'^\s*$', np.nan, regex = True)
df["ane"] = df["ane"].str.strip('\t')

filterClass = df.loc[df["classification"] == "ckd"]
countClass = filterClass["classification"].count()
print("Count of ckd for classification is " + str(countClass))

df["classification"] = df["classification"].fillna("notckd")
df["classification"] = df["classification"].replace(r'^\s*$', np.nan, regex = True)
df["classification"] = df["classification"].str.strip('\t')
```
