### 6. From the Cleveland Heart Disease dataset:

<br>
<font color=blue>
1. calculate the mean and variance for the maximum heart rate achive during the exercise test, <br>
2. then generate 10000 boostraps and calculate the average mean, variance and 95% Confidence interval for each of the bootstraps samples. <br>
3. Produce density plots of the mean bootstraps.<br> 
4. In one of two sentences write an overall conclusion of applying the bootstrap to increase the precision to this dataset.<br></font>

```r
df <- read.csv("processed.cleveland.csv", header = T)
exercise = df$X150.0
mean_exercise = mean(exercise)
cat("mean for the maximum heart rate achive during the exercise test: ", mean_exercise, "\n")
std_exercise = sd(exercise)
cat("standard deviation for the maximum heart rate achive during the exercise test: ", std_exercise)

```

```r
install.packages("mosaic")
library(mosaic)

#install.packages("binom", repos="http://R-Forge.R-project.org")
```

```r
trellis.par.set(theme=col.mosaic())
options(digits=3)

meanconfint = function (x, sigma, level = 0.95, ...) {
  se = sigma / sqrt(length(x))
  mu = mean(x)
  z = qnorm(1 - (1 - level)/2)
  out = c(mu, mu - z * se, mu + z * se)
  names(out) = c("mean", "lower", "upper")
  return(out)
}

bootstrap1 = do(10000) * mean(resample(exercise))
bootstrap1
bootstrap2 = do(10000) * sd(resample(exercise))
bootstrap2
bootstrap3 = do(10000) * meanconfint(resample(exercise), sd(resample(exercise)))
bootstrap3

densityplot(~mean, data=bootstrap1)
```

Bootstrapping increases precision by estimating errors of the the mean, standard deviaiton and mean confidence interval of the data. It does this by sampling with replacement the sample data.
